## 比較演算子

```python
# 1. 等しい（==）
print(5 == 5)   # True
print(5 == 3)   # False

# 注意：= は代入、== は比較
```

```python
# 2. 等しくない（!=）
print(5 != 3)   # True
print(5 != 5)   # False
```

```python
# 3. 大なり・小なり
print(5 > 3)    # True（5は3より大きい）
print(5 < 3)    # False（5は3より小さくない）
```

```python
# 4. 以上・以下
print(5 >= 5)   # True（5は5以上）
print(5 <= 3)   # False（5は3以下ではない）
```

## in演算子

```python
# 5. リストに含まれるか
fruits = ['apple', 'banana', 'orange']
print('banana' in fruits)   # True
print('grape' in fruits)    # False
```

```python
# 6. 文字列に含まれるか
text = 'Hello World'
print('World' in text)  # True
print('Python' in text) # False
```

## not演算子

```python
# 7. True/Falseを反転
print(not True)   # False
print(not False)  # True
```

```python
# 8. not in（含まれていない）
fruits = ['apple', 'banana', 'orange']
print('grape' not in fruits)  # True（grapeは含まれていない）
```

## 論理演算子

```python
# 9. and（両方ともTrue）
print(True and True)    # True
print(True and False)   # False
print(False and False)  # False
```

```python
# 10. or（どちらかがTrue）
print(True or False)   # True
print(False or False)  # False
```

```python
# 11. 組み合わせ例
age = 25
is_member = True

# 20歳以上かつ会員
print(age >= 20 and is_member)  # True

# 18歳未満または会員
print(age < 18 or is_member)    # True
```
