## if文の基本

```python
# 1. if文の基本形
score = 80

if score >= 60:
    print('合格です')
```

```python
# 2. インデントが重要
score = 80

if score >= 60:
    print('合格です')      # インデントあり → if文の中
    print('おめでとう!')   # インデントあり → if文の中

print('判定終了')          # インデントなし → if文の外（常に実行）
```

## else

```python
# 3. else（条件がFalseの場合）
score = 40

if score >= 60:
    print('合格です')
else:
    print('不合格です')
```

## elif

```python
# 4. elif（複数の条件）
score = 75

if score >= 90:
    print('評価: A')
elif score >= 80:
    print('評価: B')
elif score >= 70:
    print('評価: C')
elif score >= 60:
    print('評価: D')
else:
    print('評価: F')
```

## 複数条件の組み合わせ

```python
# 5. andで複数条件
age = 25
is_member = True

if age >= 20 and is_member:
    print('割引が適用されます')
```

```python
# 6. orで複数条件
day = '土曜日'

if day == '土曜日' or day == '日曜日':
    print('週末です')
else:
    print('平日です')
```

## ネスト（入れ子）

```python
# 7. if文の中にif文
age = 25
is_student = True

if age >= 18:
    print('成人です')
    if is_student:
        print('学生割引が適用されます')
else:
    print('未成年です')
```

## 文字列メソッドを使った条件分岐

```python
# 8. startswith / endswith（文字列の判定）
filename = 'report.pdf'

if filename.endswith('.pdf'):
    print('PDFファイルです')
elif filename.endswith('.xlsx'):
    print('Excelファイルです')
else:
    print('その他のファイルです')
```

```python
# 9. find / count（文字列の検索）
text = 'Hello World'

# findは見つかった位置を返す（見つからなければ-1）
if text.find('World') != -1:
    print('Worldが含まれています')

# countは出現回数を返す
message = 'banana'
print(f'aの出現回数: {message.count("a")}回')
```

## 条件分岐の活用例

```python
# 10. データの存在チェック
data = {'name': '山田'}

email = data.get('email')
if email is not None:
    print(f'メール送信先: {email}')
else:
    print('メールアドレスが登録されていません')
```

```python
# 11. リストが空かどうかのチェック
items = []

if len(items) > 0:
    print(f'{len(items)}件のデータがあります')
else:
    print('データがありません')
```
