## Noneとは

```python
# 1. Noneは「何もない」を表す特別な値
value = None
print(value)       # None
print(type(value)) # <class 'NoneType'>
```

## Noneの判定

```python
# 2. is で判定（推奨）
value = None
print(value is None)      # True
print(value is not None)  # False
```

```python
# 3. == でも判定できるが is を使うのがPythonの作法
value = None
print(value == None)  # True（動くが非推奨）
```

## Noneの活用例

```python
# 4. 辞書のgetメソッド（キーがないとNone）
user = {'name': '山田', 'age': 25}
print(user.get('name'))   # 山田
print(user.get('email'))  # None
```

```python
# 5. 関数がreturnを持たない場合はNoneを返す
def greet(name):
    print(f'Hello, {name}!')
    # return文がない

result = greet('Mike')
print(result)  # None
```

```python
# 6. Noneチェックの例
data = None

if data is None:
    print('データがありません')
else:
    print(f'データ: {data}')
```

```python
# 7. Noneでない場合のみ処理
email = user.get('email')

if email is not None:
    print(f'メールアドレス: {email}')
else:
    print('メールアドレスが設定されていません')
```
