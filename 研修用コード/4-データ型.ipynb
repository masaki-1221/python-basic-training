```python
# 13. データ型

# --- 1. 変数の宣言と型 (Type) ---

num = 2
name = "Mike"
is_ok = True

print(num, type(num))      # 整数型（Integer）
print(name, type(name))    # 文字列型（String）
print(is_ok, type(is_ok))  # ブール型 (Boolean)：Yes/No の判定などに使用
```

```python
# --- 2. 変数の再代入 ---

num = 2
name = "Mike"

num = name 
print(num, type(num)) # Mike <class 'str'>
```

```python
# --- 3. 型変換 ※実務で重要 ---

# 文字列としての "1"
str_number = "1" 
print(str_number, type(str_number)) # 1 <class 'str'>

# 整数型に変換する
new_num = int(str_number)

print(new_num, type(new_num)) # 1 <class 'int'> になっている
```

```python
# --- 4. 型ヒント (Type Hints) ---

num: int = 10
name: str = "Mike"

# 型ヒントを無視して別の型を入れても、Pythonはエラーを出さずに実行してしまう
num: int = "Mike" 
print(type(num)) # <class 'str'> (エラーにならず型が変わる)

# --- 5. 変数名のルールとエラー ---
# 以下はエラー（SyntaxError）になるためコメントアウトしています。
# 確認する際は先頭の # を外して実行してください。

# 1. 数字から始まる変数はNG
# 1name = "Error" 

# 2. 予約語（Pythonが機能として使っている言葉）はNG
# if = 1 

# 末尾に数字や、アンダースコアはOK
name1 = "OK"
_name = "OK"













# データ型による動作の違い
print(5 + 7)      # 数値の計算（12になる）
print('5' + '7')  # 文字列の連結（57になる）
```

```python
# 14. 型変換 str
price = 100

# 【エラー】文字と数字はそのままでは連結できない
# print('バナナの価格は' + price + '円です')

# 【解決策1】型を「文字列」に変換する str() を使う
print('バナナの価格は' + str(price) + '円です')

# 【解決策2】f-strings を使う
print(f'バナナの価格は{price}円です')
```

```python
# 15. 型変換 int
price = 100
count = '3'  # 文字列の「3」

# 【エラー】文字と数字はそのままでは連結できない
total_price = price * count

# 【解決策】文字を「数字」に変換する int() を使う
# 文字のままだと計算できないので、int(count) で変換してから掛ける
total_price = price * int(count)
print(total_price)
```
