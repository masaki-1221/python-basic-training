## importの基本

```python
# 1. import モジュール名
import datetime

# モジュール名.機能名 で使う
now = datetime.datetime.now()
print(now)
```

```python
# 2. import os でファイルパス操作
import os

# 現在のディレクトリを取得
print(os.getcwd())
```

## from import

```python
# 3. from モジュール import 機能
from datetime import datetime

# モジュール名なしで使える
now = datetime.now()
print(now)
```

```python
# 4. 複数の機能をインポート
from datetime import datetime, timedelta

now = datetime.now()
tomorrow = now + timedelta(days=1)
print(f'今日: {now}')
print(f'明日: {tomorrow}')
```

## asによる別名

```python
# 5. モジュールに別名を付ける
import datetime as dt

now = dt.datetime.now()
print(now)
```
