## 組み込み関数

```python
# 1. len（長さを取得）
print(len([1, 2, 3]))      # 3
print(len('Hello'))        # 5
print(len({'a': 1, 'b': 2}))  # 2
```

```python
# 2. max / min（最大・最小）
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
print(max(numbers))  # 9
print(min(numbers))  # 1
```

```python
# 3. sum（合計）
numbers = [1, 2, 3, 4, 5]
print(sum(numbers))  # 15
```

```python
# 4. sorted（並び替え、新しいリストを返す）
numbers = [3, 1, 4, 1, 5]
sorted_numbers = sorted(numbers)
print(sorted_numbers)  # [1, 1, 3, 4, 5]
print(numbers)         # 元のリストは変わらない
```

```python
# 5. abs（絶対値）
print(abs(-5))   # 5
print(abs(3.14)) # 3.14
```

```python
# 6. round（四捨五入）
print(round(3.14159))      # 3
print(round(3.14159, 2))   # 3.14（小数点以下2桁）
```

## 標準ライブラリ

```python
# 7. datetime（日付・時刻）
from datetime import datetime, timedelta

now = datetime.now()
print(f'現在: {now}')
print(f'日付: {now.date()}')
print(f'時刻: {now.time()}')
```

```python
# 8. os（ファイルパス操作）
import os

print(os.getcwd())  # 現在のディレクトリ

# パスの結合
path = os.path.join('folder', 'file.txt')
print(path)
```

```python
# 9. json（JSONデータの処理）
import json

# 辞書 → JSON文字列
data = {'name': '山田', 'age': 25}
json_str = json.dumps(data, ensure_ascii=False)
print(json_str)

# JSON文字列 → 辞書
parsed = json.loads(json_str)
print(parsed['name'])
```

```python
# 10. time（時間操作）
import time

print('処理開始')
time.sleep(1)  # 1秒待機
print('1秒経過')
```

```python
# 11. re（正規表現）
import re

text = '電話番号: 03-1234-5678'
pattern = r'\d{2,4}-\d{2,4}-\d{4}'

match = re.search(pattern, text)
if match:
    print(f'見つかりました: {match.group()}')
```
