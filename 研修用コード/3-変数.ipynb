```python
# 7. 変数

# nameという箱に 'Mike' を入れる
name = 'Mike'

# numberという箱に 5 を入れる
number = 5

# 変数の値を出力
print(name)
print(number)

# 変数と文字列の違い
# print(name)    # 変数の中身 ('Mike') が出る
# print('name')  # 文字列 ('name') がそのまま出る
```

```python
# 8. 変数の値を更新する

# nameという変数に 'Mike' を入れる
name = 'Mike'

# 値の上書き
name = 'John'

# 変数の値を出力
print(name)    # 上書き後の ('John') が出る
```

```python
# 自分自身に足す
x = 5
x = x + 3
print(f"x + 3 の結果: {x}")

print("-" * 20)

# 省略形
# x = 5
# x += 3    # x = x + 3 と同じ意味
# print(f"x += 3 の結果: {x}")

# 引き算や掛け算も同様
# y = 10
# y *= 5  # y = y * 5 と同じ
# print(f"y *= 5 の結果: {y}")
```

```python
# 9. 変数を使う意味_その1：読みやすさ

# 変数あり：何をしてるか一目瞭然
banana_price = 100
banana_count = 3
total_price = banana_price * banana_count
print(total_price)

# 変数なし：数字の意味がわからない
print(100 * 3)
```

```python
# 10. 変数を使う意味_その2：データの再利用

# --- ケース：正方形の面積を計算したい ---

# 変数なし
# 一辺の長さが「5」から「10」に変わったら、全部書き直さないといけない
print(5 * 5)

# 変数あり
# length という変数に一辺の長さを入れる
# 一辺の長さが「10」に変わっても、lengthを書き変えるだけでOK
length = 5
print(length * length)
```
