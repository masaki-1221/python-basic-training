```python
# 12. 文字列の結合

# 1. シンプルな文字列の連結
# 「+」記号を使って文字を結合する
print('お疲れ様です' + 'Python研修へようこそ！')
```

```python
# 2. 変数を使った結合
greeting = 'こんにちは、'
name = '山田さん'
message = greeting + name + '。今日も頑張りましょう！'

print(message)
```

```python
# 3. 数字と文字列の連結（エラーになる例）
# 文字列型と数値型はそのままでは「+」で連結できません。
num = 10
print(num + '件のタスクがあります。') # TypeError: unsupported operand type(s) for +: 'int' and 'str'

# エラーを回避するには、数字を文字列に変換（str()）する必要があります。
num = 10
print(str(num) + '件のタスクがあります。')
```

```python
# 4. 【推奨】f-strings
# 実務ではこのf-stringsを使うのが最も効率的でおすすめです。
# 文字列の前に「f」をつけ、変数を {} （波括弧）で囲みます。

product = 'レポート作成'
time_limit = 15

# f-stringを使って、変数の中身を埋め込んだメッセージを作成
print(f'現在のタスクは「{product}」です。残り時間はあと{time_limit}分です。')


# f-stringを使わないと...
print('現在のタスクは「' + product + '」です。残り時間はあと' + str(time_limit) + '分です。')
```
