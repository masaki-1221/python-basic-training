## 関数を使うメリット

```python
# 関数を使うと
# 1. 同じ処理を何度も書かなくて済む
# 2. 処理に名前が付くのでコードが読みやすくなる
```

## 関数の定義

```python
# 1. 関数を定義
def greet():
    print('こんにちは！')

# 2. 関数を呼び出す
greet()
greet()  # 何度でも呼び出せる
```

## 引数

```python
# 3. 引数を1つ受け取る
def greet(name):
    print(f'こんにちは、{name}さん！')

greet('山田')
greet('田中')
```

```python
# 4. 複数の引数を受け取る
def add(a, b):
    result = a + b
    print(f'{a} + {b} = {result}')

add(3, 5)
add(10, 20)
```

## 戻り値

```python
# 5. returnで値を返す
def add(a, b):
    return a + b

result = add(3, 5)
print(result)  # 8
```

```python
# 6. 戻り値を直接使う
def add(a, b):
    return a + b

print(add(10, 20))         # 30
print(add(1, 2) + add(3, 4))  # 10
```

```python
# 7. returnがない場合はNoneを返す
def greet(name):
    print(f'こんにちは、{name}さん！')

result = greet('山田')
print(result)  # None
```

## 複数の戻り値

```python
# 8. 複数の値を返す（タプルとして）
def get_name_and_age():
    name = '山田'
    age = 25
    return name, age

result = get_name_and_age()
print(result)        # ('山田', 25)
print(type(result))  # <class 'tuple'>
```

```python
# 9. 複数の変数で受け取る
def get_name_and_age():
    return '山田', 25

name, age = get_name_and_age()
print(name)  # 山田
print(age)   # 25
```

## 生成AIのコードでの関数

```python
# 10. よくあるパターン
def calculate_tax(price, tax_rate=0.1):
    """税込み価格を計算する"""
    return price * (1 + tax_rate)

# 何を受け取り、何を返すか確認
# - 引数: price（価格）, tax_rate（税率、デフォルト10%）
# - 戻り値: 税込み価格

print(calculate_tax(1000))      # 1100.0
print(calculate_tax(1000, 0.08)) # 1080.0
```
