## 引数の型ヒント

```python
# 1. 型ヒントなしの関数
def greet(name, age):
    print(f'{name}さんは{age}歳です')

greet('田中', 25)
```

```python
# 2. 型ヒント付きの関数（引数の型を明示）
def greet(name: str, age: int):
    print(f'{name}さんは{age}歳です')

greet('田中', 25)
```

## 戻り値の型ヒント

```python
# 3. 戻り値の型を明示
def add(a: int, b: int) -> int:
    return a + b

result = add(10, 20)
print(result)
```

```python
# 4. 戻り値がない関数（Noneを返す）
def say_hello(name: str) -> None:
    print(f'こんにちは、{name}さん')

say_hello('山田')
```

## よく使う型

```python
# 5. 基本的な型
def example(
    num: int,      # 整数
    price: float,  # 浮動小数点数
    name: str,     # 文字列
    flag: bool     # 真偽値
) -> str:
    return f'{name}: {num}個, {price}円, {flag}'
```

```python
# 6. コレクション型（リスト、辞書）
def process_data(
    numbers: list,   # リスト
    user: dict       # 辞書
) -> list:
    return numbers
```

## リストの要素の型

```python
# 7. リストの中身の型まで指定
def sum_numbers(numbers: list[int]) -> int:
    return sum(numbers)

result = sum_numbers([1, 2, 3, 4, 5])
print(result)
```

```python
# 8. 辞書のキーと値の型を指定
def get_age(users: dict[str, int], name: str) -> int:
    return users.get(name, 0)

user_ages = {'田中': 25, '山田': 30}
print(get_age(user_ages, '田中'))
```

## 型ヒントの注意点

```python
# 9. 型ヒントは強制ではない（エラーにならない）
def add(a: int, b: int) -> int:
    return a + b

# 文字列を渡しても実行はできる
result = add('Hello', 'World')
print(result)  # HelloWorld
```
