## デフォルト引数

```python
# 1. デフォルト値を設定
def greet(name, greeting='こんにちは'):
    print(f'{greeting}、{name}さん！')

greet('山田')              # デフォルト値を使用
greet('田中', 'おはよう')  # 値を指定
```

```python
# 2. 複数のデフォルト引数
def create_user(name, age=20, city='東京'):
    return {'name': name, 'age': age, 'city': city}

print(create_user('山田'))
print(create_user('田中', 25))
print(create_user('佐藤', 30, '大阪'))
```

## キーワード引数

```python
# 3. 引数名を指定して呼び出す
def create_user(name, age=20, city='東京'):
    return {'name': name, 'age': age, 'city': city}

# 引数名を指定すると順番を変えられる
print(create_user(name='山田', city='大阪', age=30))

# 一部だけ指定することも可能
print(create_user('田中', city='名古屋'))
```

## Docstring

```python
# 4. Docstringの書き方
def calculate_tax(price, tax_rate=0.1):
    """
    税込み価格を計算する関数
    
    Args:
        price: 税抜き価格
        tax_rate: 税率（デフォルト: 0.1）
    
    Returns:
        税込み価格
    """
    return price * (1 + tax_rate)

print(calculate_tax(1000))
```

```python
# 5. Docstringを確認（help関数）
help(calculate_tax)
```

```python
# 6. Docstringを確認（?を使う方法）
# Jupyter Notebookでは ? で確認できる
calculate_tax?
```
