```python
# 1. 数値を出力（クォーテーション不要）
print(3)
```

```python
# 2. 四則演算
print(3 + 7)   # 足し算
print(7 - 3)   # 引き算
print(3 * 7)   # 掛け算（アスタリスク）
print(3 / 2)   # 割り算（スラッシュ）
```

```python
# 3. その他の演算
print(7 % 3)   # 割り算の余り（パーセント）
print(16 // 3) # 小数点以下切り捨て（ダブルスラッシュ）
```

```python
# 4. クォーテーションで囲むと文字列になる
print('3 + 7')  # 計算されずにそのまま表示
```
