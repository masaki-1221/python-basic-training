## 文字列の結合

```python
# 1. 文字列の結合（プラス記号）
print('こんにちは、' + '山田さん')
```

```python
# 2. 文字列の繰り返し（アスタリスク）
print('Python' * 3)
print('-' * 20)
```

## インデックス

```python
# 3. インデックス（位置は0から数える）
text = 'Python'
print(text[0])  # P（1文字目）
print(text[1])  # y（2文字目）
print(text[2])  # t（3文字目）
```

```python
# 4. マイナスのインデックス（後ろから数える）
text = 'Python'
print(text[-1])  # n（最後の文字）
print(text[-2])  # o（最後から2番目）
```

## スライス

```python
# 5. スライス（範囲を指定して切り出す）
text = 'Python'
print(text[0:3])  # Pyt（0番目から2番目まで）
print(text[2:5])  # tho（2番目から4番目まで）
```

```python
# 6. スライスの省略
text = 'Python'
print(text[:3])   # Pyt（最初から2番目まで）
print(text[3:])   # hon（3番目から最後まで）
```

## 文字列メソッド

```python
# 7. upper / lower（大文字・小文字変換）
text = 'Hello World'
print(text.upper())  # HELLO WORLD
print(text.lower())  # hello world
```

```python
# 8. strip（前後の空白を削除）
text = '   Hello World   '
print(text.strip())  # Hello World
```

```python
# 9. split（文字列を分割してリストにする）
text = 'apple,banana,orange'
print(text.split(','))  # ['apple', 'banana', 'orange']
```

```python
# 10. replace（文字列を置換）
text = 'Hello World'
print(text.replace('World', 'Python'))  # Hello Python
```

## f-strings

```python
# 11. f-strings（変数を埋め込む）
name = '山田'
age = 25
print(f'私の名前は{name}です。{age}歳です。')
```

```python
# 12. f-stringsを使わない場合（比較用）
name = '山田'
age = 25
print('私の名前は' + name + 'です。' + str(age) + '歳です。')
```
