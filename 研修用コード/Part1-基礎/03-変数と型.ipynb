## 変数とは

```python
# 1. 変数の宣言と出力
name = 'Mike'
number = 5

print(name)
print(number)
```

```python
# 2. 変数と文字列の違い
name = 'Mike'
print(name)    # 変数の中身（Mike）
print('name')  # 文字列（name）
```

## 変数の更新

```python
# 3. 値の上書き
name = 'Mike'
name = 'John'
print(name)  # John
```

```python
# 4. 自分自身を使った更新
x = 5
x = x + 3
print(x)  # 8
```

```python
# 5. 省略形
x = 5
x += 3   # x = x + 3 と同じ
print(x)  # 8

y = 10
y *= 2   # y = y * 2 と同じ
print(y)  # 20
```

## 変数を使うメリット

```python
# 6. 読みやすさ（可読性）
banana_price = 100
banana_count = 3
total_price = banana_price * banana_count
print(total_price)  # 何の計算か分かりやすい

# 変数なし（何の計算か分からない）
print(100 * 3)
```

```python
# 7. 再利用しやすさ
length = 5
print(length * length)  # lengthを変えれば全体に反映
```

## 変数名のルール

```python
# 8. 良い例：英単語、アンダースコアで区切る
user_name = 'Mike'
total_price = 500

# 非推奨：ローマ字、日本語
# namae = 'Mike'
# 名前 = 'Mike'
```

## データ型

```python
# 10. 4つの基本的な型
num_int = 10          # int（整数型）
num_float = 3.14      # float（浮動小数点型）
text = 'Hello'        # str（文字列型）
is_active = True      # bool（ブール型）

print(type(num_int))
print(type(num_float))
print(type(text))
print(type(is_active))
```

```python
# 11. 型による動作の違い
print(5 + 7)      # 数値の計算 → 12
print('5' + '7')  # 文字列の結合 → 57
```

## 型変換

```python
# 12. 文字列と数値の結合（エラー）
price = 100
# print('価格は' + price + '円です')  # TypeError
```

```python
# 13. str()で型変換
price = 100
print('価格は' + str(price) + '円です')
```

```python
# 14. f-stringsなら自動変換（次のセクションで詳しく学びます）
price = 100
print(f'価格は{price}円です')
```

```python
# 15. int() / float() で数値に変換
str_num = '123'
num = int(str_num)
print(num + 100)  # 223

str_float = '3.14'
num_float = float(str_float)
print(num_float * 2)  # 6.28
```
