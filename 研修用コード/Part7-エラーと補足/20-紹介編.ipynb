## リスト内包表記

```python
# 1. 通常のfor文でリストを作る
squares = []
for i in range(5):
    squares.append(i ** 2)
print(squares)  # [0, 1, 4, 9, 16]
```

```python
# 2. リスト内包表記で1行に
squares = [i ** 2 for i in range(5)]
print(squares)  # [0, 1, 4, 9, 16]
```

```python
# 3. 条件付きリスト内包表記
# 偶数だけを抽出
evens = [i for i in range(10) if i % 2 == 0]
print(evens)  # [0, 2, 4, 6, 8]
```

## 三項演算子

```python
# 4. 通常のif-else
score = 75
if score >= 60:
    result = '合格'
else:
    result = '不合格'
print(result)
```

```python
# 5. 三項演算子で1行に
score = 75
result = '合格' if score >= 60 else '不合格'
print(result)
```

## アンパック

```python
# 6. リストやタプルのアンパック
point = (10, 20, 30)
x, y, z = point
print(x)  # 10
print(y)  # 20
print(z)  # 30
```

```python
# 7. for文でのアンパック
points = [(1, 2), (3, 4), (5, 6)]

for x, y in points:
    print(f'x={x}, y={y}')
```

```python
# 8. 辞書のitemsでのアンパック
user = {'name': '山田', 'age': 25}

for key, value in user.items():
    print(f'{key}: {value}')
```

## ラムダ式

```python
# 9. 通常の関数定義
def double(x):
    return x * 2

print(double(5))  # 10
```

```python
# 10. ラムダ式で1行に
double = lambda x: x * 2

print(double(5))  # 10
```

```python
# 11. sortedのキーとしてよく使われる
users = [
    {'name': '田中', 'age': 30},
    {'name': '山田', 'age': 25},
    {'name': '佐藤', 'age': 35}
]

# 年齢順にソート
sorted_users = sorted(users, key=lambda u: u['age'])
for user in sorted_users:
    print(user)
```
