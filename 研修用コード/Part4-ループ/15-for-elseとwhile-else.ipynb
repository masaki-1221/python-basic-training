## for-elseの基本

```python
# 1. ループが最後まで完了した場合
fruits = ['apple', 'banana', 'orange']

for fruit in fruits:
    print(fruit)
else:
    print('全ての果物を処理しました')
```

## breakした場合

```python
# 2. breakした場合、elseは実行されない
fruits = ['apple', 'banana', 'orange']

for fruit in fruits:
    print(fruit)
    if fruit == 'banana':
        print('バナナを見つけた！')
        break
else:
    print('全ての果物を処理しました')  # これは実行されない
```

## for-elseの使いどころ

```python
# 3. 探し物が見つからなかった場合の処理
numbers = [1, 3, 5, 7, 9]

for num in numbers:
    if num % 2 == 0:  # 偶数を探す
        print(f'偶数が見つかりました: {num}')
        break
else:
    print('偶数は見つかりませんでした')
```

```python
# 4. 探し物が見つかった場合
numbers = [1, 3, 4, 7, 9]

for num in numbers:
    if num % 2 == 0:  # 偶数を探す
        print(f'偶数が見つかりました: {num}')
        break
else:
    print('偶数は見つかりませんでした')  # これは実行されない
```

## while-else

```python
# 5. while-elseの例
count = 0

while count < 3:
    print(count)
    count += 1
else:
    print('ループが正常に完了しました')
```

```python
# 6. while-elseでbreakした場合
count = 0

while count < 5:
    print(count)
    if count == 2:
        print('2でbreak')
        break
    count += 1
else:
    print('ループが正常に完了しました')  # これは実行されない
```
