## for文の基本

```python
# 1. リストの要素を順番に処理
fruits = ['apple', 'banana', 'orange']

for fruit in fruits:
    print(fruit)
```

```python
# 2. 文字列の文字を順番に処理
for char in 'Python':
    print(char)
```

## range関数

```python
# 3. range(n): 0からn-1まで
for i in range(5):
    print(i)  # 0, 1, 2, 3, 4
```

```python
# 4. range(start, end): startからend-1まで
for i in range(1, 6):
    print(i)  # 1, 2, 3, 4, 5
```

```python
# 5. range(start, end, step): ステップを指定
for i in range(0, 10, 2):
    print(i)  # 0, 2, 4, 6, 8
```

## enumerate関数

```python
# 6. インデックスと要素を同時に取得
fruits = ['apple', 'banana', 'orange']

for i, fruit in enumerate(fruits):
    print(f'{i}: {fruit}')
```

```python
# 7. 開始番号を指定
fruits = ['apple', 'banana', 'orange']

for i, fruit in enumerate(fruits, start=1):
    print(f'{i}番目: {fruit}')
```

## 辞書のループ

```python
# 8. キーを取得
user = {'name': '山田', 'age': 25, 'city': '東京'}

for key in user:
    print(key)
```

```python
# 9. キーと値を同時に取得
user = {'name': '山田', 'age': 25, 'city': '東京'}

for key, value in user.items():
    print(f'{key}: {value}')
```

## ネスト（入れ子）

```python
# 10. 九九の表（一部）
for i in range(1, 4):
    for j in range(1, 4):
        print(f'{i} x {j} = {i * j}')
    print('---')
```
