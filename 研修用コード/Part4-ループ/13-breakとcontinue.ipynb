## break

```python
# 1. breakでループを終了
fruits = ['apple', 'banana', 'orange', 'grape']

for fruit in fruits:
    if fruit == 'orange':
        print('orangeが見つかりました！')
        break  # ここでループ終了
    print(fruit)
```

```python
# 2. 検索でbreakを使う例
numbers = [1, 3, 5, 7, 8, 9, 11]

for num in numbers:
    if num % 2 == 0:  # 偶数を発見
        print(f'最初の偶数: {num}')
        break
```

## continue

```python
# 3. continueで今回だけスキップ
for num in range(1, 11):
    if num % 3 == 0:  # 3の倍数はスキップ
        continue
    print(num)
```

```python
# 4. 特定の条件を除外して処理
items = ['apple', '', 'banana', '', 'orange']

for item in items:
    if item == '':  # 空文字はスキップ
        continue
    print(f'処理: {item}')
```

## breakとcontinueの違い

```python
# 5. breakの動作確認
print('--- break ---')
for i in range(5):
    if i == 3:
        break  # ループ自体を終了
    print(i)
print('ループ終了')
```

```python
# 6. continueの動作確認
print('--- continue ---')
for i in range(5):
    if i == 3:
        continue  # 今回だけスキップ
    print(i)
print('ループ終了')
```

## while文での使用

```python
# 7. while True + break のパターン
count = 0

while True:
    print(f'カウント: {count}')
    count += 1
    
    if count >= 5:
        break  # 条件を満たしたら終了
```
