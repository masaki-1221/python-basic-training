## while文の基本

```python
# 1. 条件がTrueの間、繰り返す
count = 0

while count < 5:
    print(count)
    count += 1  # カウントを増やす
```

```python
# 2. ループ後のカウントを確認
count = 0

while count < 5:
    count += 1

print(f'ループ終了後のcount: {count}')
```

## for文との使い分け

```python
# 4. for文：繰り返す回数が決まっている場合
for i in range(5):
    print(f'処理{i}回目')
```

```python
# 5. while文：条件が満たされるまで繰り返す場合
total = 0
num = 1

while total < 100:
    total += num
    num += 1

print(f'合計が100を超えたのは {num - 1} まで足したとき')
print(f'合計: {total}')
```
