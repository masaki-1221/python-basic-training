## zip関数の基本

```python
# 1. インデックスを使う方法（従来のやり方）
names = ['田中', '山田', '佐藤']
ages = [25, 30, 35]

for i in range(len(names)):
    print(f'{names[i]}さんは{ages[i]}歳')
```

```python
# 2. zip関数を使う方法（シンプル）
names = ['田中', '山田', '佐藤']
ages = [25, 30, 35]

for name, age in zip(names, ages):
    print(f'{name}さんは{age}歳')
```

## zip関数の仕組み

```python
# 3. zipの中身を確認
names = ['田中', '山田', '佐藤']
ages = [25, 30, 35]

zipped = list(zip(names, ages))
print(zipped)  # [('田中', 25), ('山田', 30), ('佐藤', 35)]
```

```python
# 4. 3つ以上のリストも同時に処理
names = ['田中', '山田', '佐藤']
ages = [25, 30, 35]
cities = ['東京', '大阪', '福岡']

for name, age, city in zip(names, ages, cities):
    print(f'{name}さん: {age}歳, {city}在住')
```

## リストの長さが異なる場合

```python
# 5. 短い方に合わせる
names = ['田中', '山田', '佐藤', '鈴木']
ages = [25, 30]

for name, age in zip(names, ages):
    print(f'{name}さんは{age}歳')
# 佐藤と鈴木は処理されない
```

## 辞書の作成

```python
# 6. zipを使って辞書を作成
keys = ['name', 'age', 'city']
values = ['田中', 25, '東京']

user = dict(zip(keys, values))
print(user)  # {'name': '田中', 'age': 25, 'city': '東京'}
```
