## 集合型とは

```python
# 1. 集合型の作成（重複は自動的に除かれる）
fruits = {'apple', 'banana', 'apple', 'orange', 'banana'}
print(fruits)  # 重複が除かれた状態で表示
```

```python
# 2. 型の確認
fruits = {'apple', 'banana', 'orange'}
print(type(fruits))  # <class 'set'>
```

## 集合型の作成

```python
# 3. リストからsetを作成（重複削除によく使う）
my_list = [1, 2, 2, 3, 3, 3, 4]
unique_values = set(my_list)
print(unique_values)  # {1, 2, 3, 4}
```

```python
# 4. 空の集合を作る場合の注意
empty_set = set()     # 正しい
empty_dict = {}       # これは空の辞書になる

print(type(empty_set))   # <class 'set'>
print(type(empty_dict))  # <class 'dict'>
```

## 集合の操作

```python
# 5. 要素の追加
fruits = {'apple', 'banana'}
fruits.add('orange')
print(fruits)
```

```python
# 6. 要素の削除
fruits = {'apple', 'banana', 'orange'}
fruits.remove('banana')
print(fruits)
```

## 集合演算

```python
# 7. 和集合（どちらかに含まれる要素）
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)       # {1, 2, 3, 4, 5}
print(a.union(b))  # 同じ結果
```

```python
# 8. 積集合（両方に共通する要素）
a = {1, 2, 3}
b = {3, 4, 5}

print(a & b)              # {3}
print(a.intersection(b))  # 同じ結果
```

```python
# 9. 差集合（一方にだけ含まれる要素）
a = {1, 2, 3}
b = {3, 4, 5}

print(a - b)            # {1, 2}
print(a.difference(b))  # 同じ結果
```

## 集合型の使いどころ

```python
# 10. リストの重複を削除
purchases = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
unique_fruits = set(purchases)
print(unique_fruits)  # ユニークな果物だけ
```

```python
# 11. 共通要素を見つける
my_friends = {'Alice', 'Bob', 'Charlie'}
your_friends = {'Bob', 'David', 'Eve'}

common_friends = my_friends & your_friends
print(common_friends)  # {'Bob'}
```
