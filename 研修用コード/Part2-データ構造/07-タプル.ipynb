## タプルの作成

```python
# 1. タプルの作成（丸括弧）
point = (10, 20)
print(point)
```

```python
# 2. インデックスで要素を取得
point = (10, 20, 30)
print(point[0])  # 10
print(point[1])  # 20
```

## タプルとリストの違い

```python
# 3. リストは変更できる
my_list = [1, 2, 3]
my_list[0] = 100
print(my_list)  # [100, 2, 3]
```

```python
# 4. タプルは変更できない（エラー）
my_tuple = (1, 2, 3)
# my_tuple[0] = 100  # TypeError
```

## タプルの用途

```python
# 5. 関数から複数の値を返す
def get_position():
    x = 10
    y = 20
    return (x, y)

result = get_position()
print(result)     # (10, 20)
print(result[0])  # 10
print(result[1])  # 20
```

```python
# 6. 複数の変数で受け取る
def get_position():
    return (10, 20)

x, y = get_position()
print(x)  # 10
print(y)  # 20
```

```python
# 7. 辞書のitemsはタプルを返す
user = {'name': '山田', 'age': 25}

for item in user.items():
    print(item)        # タプルとして表示
    print(type(item))  # <class 'tuple'>
```
