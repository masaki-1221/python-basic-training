## 辞書の作成

```python
# 1. 辞書の作成（キー: 値）
user = {
    'name': '山田',
    'age': 25,
    'department': '営業部'
}
print(user)
```

## 値の取得

```python
# 2. 角括弧でキーを指定
user = {'name': '山田', 'age': 25}
print(user['name'])  # 山田
print(user['age'])   # 25
```

```python
# 3. 存在しないキーを指定するとエラー
user = {'name': '山田', 'age': 25}
# print(user['email'])  # KeyError
```

```python
# 4. getメソッド（キーがなくてもエラーにならない）
user = {'name': '山田', 'age': 25}
print(user.get('name'))   # 山田
print(user.get('email'))  # None
print(user.get('email', '未設定'))  # 未設定（デフォルト値）
```

## 値の追加・変更

```python
# 5. 新しいキーを追加
user = {'name': '山田', 'age': 25}
user['email'] = 'yamada@example.com'
print(user)
```

```python
# 6. 既存のキーを更新
user = {'name': '山田', 'age': 25}
user['age'] = 26
print(user)
```

## 辞書のメソッド

```python
# 7. keys（全てのキーを取得）
user = {'name': '山田', 'age': 25, 'department': '営業部'}
print(user.keys())
```

```python
# 8. values（全ての値を取得）
user = {'name': '山田', 'age': 25, 'department': '営業部'}
print(user.values())
```

```python
# 9. items（キーと値のペアを取得）
user = {'name': '山田', 'age': 25, 'department': '営業部'}
print(user.items())
```
