## リストの作成

```python
# 1. リストの作成
fruits = ['apple', 'banana', 'orange']
print(fruits)
```

```python
# 2. 異なる型を混在させることも可能
mixed = ['apple', 100, True, 3.14]
print(mixed)
```

## 文字列からリストを作る（split）

```python
# 3. splitで文字列をリストに変換
text = 'apple,banana,orange'
fruits = text.split(',')
print(fruits)  # ['apple', 'banana', 'orange']

# スペース区切りの場合
sentence = 'Hello World Python'
words = sentence.split(' ')
print(words)  # ['Hello', 'World', 'Python']
```

## インデックスとスライス

```python
# 4. インデックスで要素を取得
fruits = ['apple', 'banana', 'orange']
print(fruits[0])  # apple
print(fruits[1])  # banana
print(fruits[-1]) # orange（最後の要素）
```

```python
# 5. スライスで範囲を取得
numbers = [0, 1, 2, 3, 4, 5]
print(numbers[1:4])  # [1, 2, 3]
print(numbers[:3])   # [0, 1, 2]
print(numbers[3:])   # [3, 4, 5]
```

## 要素の変更

```python
# 6. 要素を上書き
fruits = ['apple', 'banana', 'orange']
fruits[1] = 'grape'
print(fruits)  # ['apple', 'grape', 'orange']
```

## リストのメソッド

```python
# 7. append（末尾に追加）
fruits = ['apple', 'banana']
fruits.append('orange')
print(fruits)  # ['apple', 'banana', 'orange']
```

```python
# 8. insert（指定位置に挿入）
fruits = ['apple', 'orange']
fruits.insert(1, 'banana')  # 1番目に挿入
print(fruits)  # ['apple', 'banana', 'orange']
```

```python
# 9. remove（値を指定して削除）
fruits = ['apple', 'banana', 'orange']
fruits.remove('banana')
print(fruits)  # ['apple', 'orange']
```

```python
# 10. pop（位置を指定して取り出し）
fruits = ['apple', 'banana', 'orange']
removed = fruits.pop(1)  # 1番目を取り出す
print(removed)  # banana
print(fruits)   # ['apple', 'orange']
```

```python
# 11. sort（並び替え）
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
numbers.sort()
print(numbers)  # [1, 1, 2, 3, 4, 5, 6, 9]
```

## リストの長さ

```python
# 12. len関数で要素数を取得
fruits = ['apple', 'banana', 'orange']
print(len(fruits))  # 3
```
